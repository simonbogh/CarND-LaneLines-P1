## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    # slope left: positive = right line
    # slope right: negative = left line
    leftLineXs = []
    leftLineYs = []
    rightLineXs = []
    rightLineYs = []

    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = ((y2-y1)/(x2-x1))
            if slope > 0: # right line
                rightLineXs.append(x1)
                rightLineYs.append(y1)
                rightLineXs.append(x2)
                rightLineYs.append(y2)
            else: # left line
                leftLineXs.append(x1)
                leftLineYs.append(y1)
                leftLineXs.append(x2)
                leftLineYs.append(y2)

    # Check if there are no points to fit a new line
    if len(rightLineXs) == 0 or len(leftLineXs) == 0:
        return

    # Fit a polynominal to the points and find slope (a) and intercept (b) of the fitted line
    rightLineSlope, rightLineIntercept = np.polyfit(rightLineXs, rightLineYs, 1)
    leftLineSlope, leftLineIntercept = np.polyfit(leftLineXs, leftLineYs, 1)

    # Find the max lenght to draw
    maxX = img.shape[1]
    maxY = img.shape[0]

    # New right line
    right_y1 = int(maxY)
    right_x1 = int((right_y1 - rightLineIntercept) / rightLineSlope) #  y = a*x + b => x = (y-b)/a
    right_y2 = int(maxY/2 + 50)
    right_x2 = int((right_y2 - rightLineIntercept) / rightLineSlope) #  y = a*x + b => x = (y-b)/a
    # Draw line
    cv2.line(img, (right_x1, right_y1), (right_x2, right_y2), color, thickness)

    # New left line
    left_y1 = int(maxY)
    left_x1 = int((left_y1 - leftLineIntercept) / leftLineSlope) #  y = a*x + b => x = (y-b)/a
    left_y2 = int(maxY/2 + 50)
    left_x2 = int((left_y2 - leftLineIntercept) / leftLineSlope) #  y = a*x + b => x = (y-b)/a
    # Draw line
    cv2.line(img, (left_x1, left_y1), (left_x2, left_y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images_output directory.
import os
test_images_directory = "test_images/"
test_images_names = os.listdir(test_images_directory)

test_images = []
test_images_gray = []
# Load and display images
for image_name in test_images_names:
    # Load image
    img = mpimg.imread(test_images_directory+image_name)
    test_images.append(img)
    test_images_gray.append(grayscale(img))
    #plt.figure()
    #plt.title(image_name)
    #plt.imshow(img)

# Display test image
img_idx = 1
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
ax1.set_title('Original')
ax1.imshow(test_images[img_idx])
ax2 = fig.add_subplot(1,2,2)
ax2.set_title('Gray')
ax2.imshow(test_images_gray[img_idx], cmap='gray')
````

````python
# Add gaussian blur
def apply_blur(img):
    kernel_size = 15 # Kernel size must be an odd number so it becomes 3x3, 5x5, etc.
    image_blur = gaussian_blur(img, kernel_size)
    return image_blur

test_image_color = test_images[img_idx]
test_image = test_images_gray[img_idx]
test_image_blur = apply_blur(test_image)
plt.imshow(test_image_blur, cmap='gray')
````

````python
# Canny edge detection
# Define parameters for Canny and apply
def apply_canny(img):
    low_threshold = 10
    high_threshold = 110
    img_edges = canny(img, low_threshold, high_threshold)
    return img_edges
    
test_image_edges = apply_canny(test_image_blur)
plt.imshow(test_image_edges, cmap='gray')
````

````python
# Keep region of interest (ROI)
# Find size of image
def apply_roi(img):
    ysize = img.shape[0]
    xsize = img.shape[1]
    # Define three points for region to mask
    left_bottom = [120, ysize]
    right_bottom = [900, ysize]
    apex = [xsize/2, 280]
    # Define an array with the points
    region = [np.array([left_bottom, apex, right_bottom], np.int)]
    # Mask the image
    image_masked = region_of_interest(img, region)
    return image_masked

# Mask canny image
test_image_masked = apply_roi(test_image_edges)
# Plot the masked image
plt.figure()
plt.imshow(test_image_masked, cmap='gray')
````

````python
# Define the Hough transform parameters
# Make a blank the same size as our image to draw on.
# rho and theta are the distance and angular resolution of our grid in Hough space.
#  - rho: pixels
#  - theta: radians
#  - threshold: specifies the minimum number of votes (intersections in a given grid cell) a candidate line needs to have to make it into the output
#  - min_line_len: minimum length of a line in pixels that we accept in the output
#  - max_line_gap: maximum distance in pixels between segments that we allow to be connected into a single line 
def apply_hough(img):
    rho = 1
    theta = np.pi/180
    threshold = 8
    min_line_len = 18
    max_line_gap = 1

    # Run Hough on edge detected image and draw lines
    img_with_lines = hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap)
    return img_with_lines

# Find lines in masked image
test_img_with_lines = apply_hough(test_image_masked)
# Plot image with lines found
plt.figure()
plt.title('Hough lines')
plt.imshow(test_img_with_lines)
````

````python
# Overlay the found lines on the original image
def apply_lines_to_original(img_lines, org_img):
    """
    img_lines: image with Hough lines
    org_img: original color image
    """
    image_weighted = weighted_img(img_lines, org_img, =0.8, =1., =0.)
    return image_weighted

# Overlay lines on the original image
test_final_image = apply_lines_to_original(test_img_with_lines, test_image_color)
# Plot the original image with lines
plt.figure()
plt.title('Final image with lines')
plt.imshow(test_final_image)
````
